## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
study_results
mouse_metadata

# Combine the data into a single dataset

study_df = pd.merge(study_results, mouse_metadata, on="Mouse ID")

# Display the data table for preview
study_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
study_df_grouped = study_df.groupby(["Drug Regimen"])
                    
tumor_mean = study_df_grouped["Tumor Volume (mm3)"].mean()        
tumor_median = study_df_grouped["Tumor Volume (mm3)"].median()
tumor_var = study_df_grouped["Tumor Volume (mm3)"].var()
tumor_std = study_df_grouped["Tumor Volume (mm3)"].std()
tumor_sem = study_df_grouped["Tumor Volume (mm3)"].sem()

tumor_stats_df = pd.DataFrame(
                {"Average Tumor Size":tumor_mean,
                 "Median Tumor Size":tumor_median,
                 "Tumor Size Variance":tumor_var, 
                 "Tumor Size StdDev":tumor_std, 
                 "StdErr of Mean Tumor Size":tumor_sem})

tumor_stats_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
# study_df_grouped = study_df.groupby(["Drug Regimen"])
                    
tumor_stats_df2 = study_df.groupby(['Drug Regimen']).\
                agg({"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]})

                             
tumor_stats_df2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

study_df_new = study_df.groupby(["Mouse ID"]).agg({"Timepoint": ["max"]})
study_df_new.columns = study_df_new.columns.droplevel(1)




# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

study_df_new = pd.merge(study_df_new, study_df, on=["Mouse ID", "Timepoint"], how="left")
study_df_new
```

```python
# Put treatments into a list for for loop (and later for plot labels)

treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]


# Create empty list to fill with tumor vol data (for plotting)
final_tumor_sizes = []

for drug in treatments:
    final_tumor_sizes.append(study_df_new["Tumor Volume (mm3)"][(study_df_new["Drug Regimen"] == drug)])
    

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

lowerq = [0,0,0,0]
upperq = [0,0,0,0]
iqr = [0,0,0,0]
upper_bound = [0,0,0,0]
lower_bound = [0,0,0,0]

outliers = []

for x in range(4):
    quartiles = final_tumor_sizes[x].quantile([.25,.5,.75])
    lowerq[x] = quartiles[0.25]
    upperq[x] = quartiles[0.75]
    iqr[x] = upperq[x]-lowerq[x]
    lower_bound[x] = lowerq[x] - (1.5*iqr[x])
    upper_bound[x] = upperq[x] + (1.5*iqr[x])
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset
    # Determine outliers using upper and lower bounds
    outliers.append(study_df_new["Tumor Volume (mm3)"][(study_df_new["Drug Regimen"] == treatments[x])&
                                 ((study_df_new["Tumor Volume (mm3)"] > upper_bound[x]) |
                                (study_df_new["Tumor Volume (mm3)"] < lower_bound[x]))].values)
                                 
for x in range(4):
    print(f"Outliers for {treatments[x]} regimen are below {lower_bound[x]} or above {upper_bound[x]}")
    print(f"Outliers found: {outliers[x]}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volumes by Drug Regimen')
ax1.set_ylabel('Tumor Volume (mm3)')
ax1.boxplot(final_tumor_sizes)
ax1.set_xticklabels(treatments)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
mouse = "b128"
mouse_df = study_df[["Timepoint", "Tumor Volume (mm3)"]][(study_df["Mouse ID"] == mouse)]

mouse_df

x_axis = mouse_df["Timepoint"].values
y_data = mouse_df["Tumor Volume (mm3)"].values

plt.plot(x_axis, y_data)
plt.title(f"Tumor Volume vs Time for mouse {mouse}")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Timepoint")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin_df = study_df[["Mouse ID", "Weight (g)", "Tumor Volume (mm3)"]]\
                        [(study_df["Drug Regimen"] == "Capomulin")]
capomulin_df
capomulin_df_grouped = capomulin_df.groupby(["Mouse ID"]).agg({"Weight (g)": ["max"],\
                                                            "Tumor Volume (mm3)": ["mean"]})

capomulin_df_grouped.columns = capomulin_df_grouped.columns.droplevel(1)
capomulin_df_grouped 

x_values = capomulin_df_grouped["Weight (g)"].values
y_values = capomulin_df_grouped["Tumor Volume (mm3)"].values

plt.scatter(x_values, y_values)
plt.title("Average Tumor Volume vs Weight for Capomulin treated mice")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(19,35),fontsize=15,color="red")
plt.title("Average Tumor Volume vs Weight for Capomulin treated mice")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
